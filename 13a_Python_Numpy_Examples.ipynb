# Temperature Analysis with NumPy and Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt  # This is a plotting library.  We'll learn more about this later
```

```python
# Step 1: Data Setup
# Each row corresponds to a city and each column to a month (Jan to Dec)

temperature_data = np.array([
    [3.1, 4.0, 8.3, 13.4, 18.3, 22.0, 25.1, 24.8, 20.0, 14.5, 8.9, 4.2],     # City A
    [-5.2, -2.8, 2.4, 10.2, 17.5, 22.8, 25.4, 24.1, 18.3, 10.0, 2.2, -3.3],  # City B
    [10.2, 11.5, 14.0, 17.8, 21.0, 24.5, 27.2, 27.1, 23.5, 19.0, 14.3, 11.0] # City C
])

city_names = np.array(['City A', 'City B', 'City C'])
month_names = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                        'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
```

```python
# Step 2: Compute Yearly Averages
yearly_avg = temperature_data.mean(axis=1)

for i, avg in enumerate(yearly_avg):
    print(f"{city_names[i]} - Yearly Average Temperature: {avg:.1f} °C")
```

```python
# Step 3: Identify Hottest and Coldest Months
hottest_months = temperature_data.argmax(axis=1)  # location of max
coldest_months = temperature_data.argmin(axis=1)  # location of min

for i in range(len(city_names)):
    print(f"\n{city_names[i]}:")
    print(f"  Hottest Month: {month_names[hottest_months[i]]} ({temperature_data[i, hottest_months[i]]} °C)")
    print(f"  Coldest Month: {month_names[coldest_months[i]]} ({temperature_data[i, coldest_months[i]]} °C)")
```

```python
# Step 4: Plotting Temperature Trends
plt.figure(figsize=(10, 6))
for i in range(len(city_names)):
    plt.plot(month_names, temperature_data[i], marker='o', label=city_names[i])

plt.title('Monthly Average Temperatures by City')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Student Exam Score

```python
# Step 1: Data Setup
# Each row corresponds to a city and each column to a month (Jan to Dec)

exam_scores_data = np.array([
                              [85,65,88,74,68,94,87],
                              [95,93,98,85,84,91,75],
                              [75,66,87,72,90,81,67]
])

students = np.array(['Student 1', 'Student 2', 'Student 3'])
subject = np.array(['Math','Science','CS','History','English','Spanish','Economics'])
```

```python
# Average per student
average_per_student = exam_scores_data.mean(axis=1)

for i, avg in enumerate(average_per_student):
    print(f"{students[i]} - Average Score: {avg:.1f}")
```

```python
# Average per subject
average_per_subject = exam_scores_data.mean(axis=0)

for i, avg in enumerate(average_per_subject):
    print(f"{subject[i]} - Average Score: {avg:.1f}")
```

```python
# Sort students based on average score
sorted_indices = np.argsort(average_per_student)[::-1]  # the [::-1] was so that the top score is first and then descending order.
sorted_students = students[sorted_indices]
print(sorted_students)
```

```python
# Plot Student Performance
# Bar Plot
x = np.arange(len(subject))  # Label locations
bar_width = 0.15

plt.figure(figsize=(10, 6))
for i in range(len(students)):
    plt.bar(x + i * bar_width, exam_scores_data[i], width=bar_width, label=students[i])

plt.xticks(x + bar_width * 2, subject)
plt.title('Exam Scores by Student')
plt.xlabel('Subject')
plt.ylabel('Score')
plt.legend()
plt.tight_layout()
plt.show()
```
